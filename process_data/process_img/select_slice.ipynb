```python
import nibabel as nib
from lungmask import mask
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Declare path of folders contain original CT image, largest slice image 
# and its segmentation image
root_img_path = '/content/drive/MyDrive/Finalterm ANN/img'
root_slice_path = '/content/drive/MyDrive/Finalterm ANN/slice'
root_seg_path = '/content/drive/MyDrive/Finalterm ANN/seg'
```

```python
# From original CT image create largest slice image and its segmentation image
for img in os.listdir(root_img_path):
    img_path = os.path.join(root_img_path, img)

    # Load data from original CT image
    img_load = nib.load(img_path)
    img_data = img_load.get_fdata()
    
    # Segment original CT image (cost much time)
    seg = mask.apply(img_data)
    
    # Select slice has max lung area and its segmentation
    lung_areas = np.sum(seg, axis=(0, 1))
    largest_slice_index = np.argmax(lung_areas)
    largest_slice = img_data[:, :, largest_slice_index]
    largest_seg = seg[:, :, largest_slice_index]
    
    # Save slice image (.nii file)
    largest_slice_nifti = nib.Nifti1Image(largest_slice, affine=img_load.affine)
    slice_path = os.path.join(root_slice_path, img.replace('.', '_selected.'))
    nib.save(largest_slice_nifti, slice_path)
    
    # Save segmentation (.nii file)
    largest_seg_nifti = nib.Nifti1Image(largest_seg, affine=img_load.affine)
    seg_path = os.path.join(root_seg_path, img.replace('.', '_seg.'))
    nib.save(largest_seg_nifti, seg_path)

    # # Display
    # fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    # ax[0].imshow(largest_slice, cmap='gray')
    # ax[0].set_title('Largest Slice')
    # ax[0].axis('off')

    # ax[1].imshow(largest_seg, cmap='gray')
    # ax[1].set_title('Segmentation')
    # ax[1].axis('off')

    # plt.tight_layout()
    # plt.show()

    # Completion notice for an initial CT image
    print(f'Sucessfully select and segment for {img}')
```
