```python
df = df[df.index.isin(id)]
```

```python
col = {'Age at Histological Diagnosis',
 'Ethnicity',
 'Gender',
 'Histology ',
 'Recurrence',
 'Smoking status'}
```

```python
df = df[col]
```

```python
df.columns
```

```python
# Standardizes, one-hot & categorically encodes
fm = []
ns = []
cs = []
for status in df["Smoking status"]:
    if status == "Former":
        fm.append(1)
        ns.append(0)
        cs.append(0)
    elif status == "Nonsmoker":
        fm.append(0)
        ns.append(1)
        cs.append(0)
    elif status == "Current":
        fm.append(0)
        ns.append(0)
        cs.append(1)
df["Former smoker"] = fm
df["Non smoker"] = ns
df["Current smoker"] = cs
df.drop('Smoking status', axis=1, inplace=True)

# Encoding & normalizing
ordinal_feats = ["Former smoker", "Non smoker", "Current smoker", "Recurrence"]
hotenc_feats = ["Gender", "Ethnicity", "Histology "]
scaled_feats = ["Age at Histological Diagnosis"]

for o in ordinal_feats:
    ordenc = OrdinalEncoder()
    df[[o]] = ordenc.fit_transform(df[[o]])

df = pd.get_dummies(df, columns=hotenc_feats)

for o in scaled_feats:
    scaler = StandardScaler()
    df[[o]] = scaler.fit_transform(df[[o]])
```

```python
label = df['Recurrence']
df = df.drop(columns=['Recurrence'])
```
