```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Read 3 data types
df_ct = pd.read_csv('/content/drive/MyDrive/Finalterm ANN/dataset/features.csv')
df_clinical = pd.read_csv('/content/drive/MyDrive/Finalterm ANN/dataset/clinical.csv')
df_genomic = pd.read_csv('/content/drive/MyDrive/Finalterm ANN/dataset/gen.csv')
```

```python
# Concate data
df_combined = pd.merge(df_ct, df_clinical, on='Case ID')
df_combined = pd.merge(df_combined, df_genomic, on='Case ID')
```

```python
df_combined
```

```python
df_combined.info()
```

```python
df_train, df_test = train_test_split(df_combined, test_size=0.1, random_state=42)
```

```python
label = pd.read_csv("/content/drive/MyDrive/Finalterm ANN/dataset/label.csv", index_col='Case ID')
```

```python
df_train = pd.merge(df_train, label, on='Case ID')
```

```python
df_test = pd.merge(df_test, label, on='Case ID')
```
