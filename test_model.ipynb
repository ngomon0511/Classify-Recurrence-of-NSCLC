```python
import pandas as pd
import numpy as np
from tensorflow import keras
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
model = keras.models.load_model("/content/drive/MyDrive/Finalterm ANN/model.h5")
```

```python
df = pd.read_csv('/content/drive/MyDrive/Finalterm ANN/dataset/data_test.csv', index_col='Case ID')
df
```

```python
X = df.drop(['Recurrence'], axis=1)
X
```

```python
y = df['Recurrence']
y
```

```python
y_pred = model.predict(X)
y_pred_labels = np.argmax(y_pred, axis=1)
```

```python
results_df = pd.DataFrame({'Case ID': X.index, 'Actual': y.values, 'Predict': y_pred_labels})
results_df['Actual'] = results_df['Actual'].replace({1: "Recurrence", 0: "Not Recurrence"})
results_df['Predict'] = results_df['Predict'].replace({1: "Recurrence", 0: "Not Recurrence"})
print(results_df)
```
